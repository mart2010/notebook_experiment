## Data Integrity QA Part-1: 

##Entity checks:
    * checks all labels and counts
    * checks Duplicate node entity (nodes with same UUID)
    * checks all disconnected nodes
    * checks nodes that do not have certain property or proprty value seems wrong



**QC: Report all Labels and entity count**

```python
%%cypher $cypher_conn

match (n)
return distinct labels(n) as labels, count(*) as nb_occurence
order by nb_occurence desc
```

**QC: Duplicate node's UUID**

```python
%%cypher $cypher_conn

match (n) 
with n.uuid as nodeUUId, count(*) as nodeCount 
where nodeCount > 1 
return nodeUUId, nodeCount
```

**QC: Disconnected nodes** 

```python
%%cypher $cypher_conn

match (n)
where NOT n -- ()
return n
```

**QC: Nodes without a 'name' property**

```python
%%cypher $cypher_conn

match (n)
where n.name is null
return n
```

**QC: Nodes for which 'name' property has tiny size**

```python
%%cypher $cypher_conn

match (n)
where size(n.name) < 3
return n
```

## Data Integrity QA Part-2:

## Relationships
    * reports relationship types occurence
    * reports relationship type by entity (Label)
    * checks mandatory relationship between specific entity 
    * checks relationship without type (impossible: Neo4J enforces a single type defined for all relationship)
    * etc...

**QC: Report relationship types occurences**

```python
%%cypher $cypher_conn

match (n)-[r]-(m) 
return type(r) as rela_type, count(*) as nb_occurence
order by nb_occurence desc
```

**QC: Report relationship type by entity**

```python
%%cypher $cypher_conn

match (n)-[r]->(m) 
return labels(n) as source, type(r) as type, labels(m) as target, count(*) as nb_occurence
order by nb_occurence desc
```

**QC: Check mandatory relationship**

```python
%%cypher $cypher_conn

//Activity without Agent ? 
match (a:ACTIVITY)
optional match (a)-[r:WAS_ASSOCIATED_WITH]->(g:AGENT) 
with a,g
where r is null
return a
```
